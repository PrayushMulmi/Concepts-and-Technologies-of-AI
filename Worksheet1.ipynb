```python
#Problem 4 - Matrix Operations

#For the following arrays:
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

#Prove the following with NumPy:

#1 - Prove : A.A^(-1) = 1

IoA = np.linalg.inv(A)

np.dot(A, IoA)

#2 - Prove A.B != B.A

print("A.B : ", np.dot(A, B))
print("B.A : ", np.dot(B, A))

#3 - Prove (A.B)^T = B^T.A^T

abdot = np.dot(A,B)
abdot_transpose = np.transpose(abdot)

atranspose = np.transpose(A)
btranspose = np.transpose(B)

btat = np.dot(btranspose, atranspose)

print("(A.B)^T : ", abdot_transpose)
print("B^T.A^T : ", btat)



#Solve the following system of linear equation using inverse methods

#2x - 3y + z = -1
#x - y + 2z = -3
#3x + y - z = 9

X = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
Y = np.array([-1, -3, 9])

#X.Z =Y
IoX = np.linalg.inv(X)
Z = np.dot(IoX, Y)
print(Z)

#{Hint : First use Numpy array to represent the equation in Matrix form. Then Solve for A.X = B}



#Now : solve the above equation using np.linalg.inv function. [Explore more about 'linalg' function of Numpy]

```
