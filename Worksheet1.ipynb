4.2 : Experimenting : How Fast is Numpy?

1. Element-wise Addition:

  • Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.

  • Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

```python
import numpy as np
from timeit import timeit

python_list = []

for i in range(1000001):
  python_list.append(i)

def add_python(python_list):
  sum_of_python_list = 0
  for i in range(1000001):
    sum_of_python_list += python_list[i]
  return sum_of_python_list

numpy_list = np.arange(1000000)

def add_numpy(numpy_list):
  sum_of_numpy_array = numpy_list.sum()
  return sum_of_numpy_array

python_time = timeit("add_python(python_list)", globals=globals(), number=100)
numpy_time = timeit("add_numpy(numpy_list)", globals=globals(), number=100)

print(f"Pure Python addition : {python_time:.6f} seconds per loop")
print(f"Numpy addition : {numpy_time:.6f} seconds per loop")

```

2. Element-wise Multiplication:

• Using Python Lists, perform element-wise multiplication of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

```python
import numpy as np
from timeit import timeit

python_list1 = []
python_list2 = []

for i in range(1000001):
  python_list1.append(i)
for j in range(1000001, 0, -1):
  python_list2.append(j)

def multiplication_python(python_list1, python_list2):
  new_python_list = []
  for i in range(1000000):
    new_python_list.append(python_list1[i] * python_list2[i])
  return new_python_list

numpy_list1 = np.arange(1000000)
numpy_list2 = np.arange(1000000)

def multiplication_numpy(numpy_list1, numpy_list2):
  return numpy_list1 * numpy_list2

python_time = timeit("multiplication_python(python_list1, python_list2)", globals=globals(), number=100)
numpy_time = timeit("multiplication_numpy(numpy_list1, numpy_list2)", globals=globals(), number=100)

print(f"Pure Python addition : {python_time:.6f} seconds per loop")
print(f"Numpy addition : {numpy_time:.6f} seconds per loop")
```

3. Dot Product:

• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

```python
import numpy as np
from timeit import timeit

python_list1 = []
python_list2 = []

for i in range(1000001):
  python_list1.append(i)
for j in range(1000001, 0, -1):
  python_list2.append(j)

def dot_python(python_list1, python_list2):
  total = 0
  for i in range(1000000):
    total += python_list1[i] * python_list2[i]
  return total

numpy_list1 = np.arange(1000000)
numpy_list2 = np.arange(1000000)

def dot_numpy(numpy_list1, numpy_list2):
  return np.dot(numpy_list1, numpy_list2)

python_time = timeit("dot_python(python_list1, python_list2)", globals=globals(), number=100)
numpy_time = timeit("dot_numpy(numpy_list1, numpy_list2)", globals=globals(), number=100)

print(f"Pure Python addition : {python_time:.6f} seconds per loop")
print(f"Numpy addition : {numpy_time:.6f} seconds per loop")
```

4. Matrix Multiplication:

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
import numpy as np
import random
from timeit import timeit

python_matrix1 = []
python_matrix2 = []

for i in range(1000):
  row = []
  for j in range(1000):
    row.append(random.random())
  python_matrix1.append(row)

for i in range(1000):
  row = []
  for j in range(1000):
    row.append(random.random())
  python_matrix2.append(row)


def matrix_python(python_matrix1, python_matrix2):
  result = []
  for i in range(1000):
    row = []
    for j in range(1000):
      res = 0
      for k in range(1000):
        res += python_matrix1[i][k] + python_matrix2[i][k]
        row.append(res)
    result.append(row)
  return result

numpy_matrix1 = np.random.rand(1000)
numpy_matrix2 = np.random.rand(1000)

def matrix_numpy(numpy_matrix1, numpy_matrix2):
  return np.dot(numpy_matrix1, numpy_matrix2)

python_time = timeit("matrix_python(python_matrix1, python_matrix2)", globals=globals(), number=100)
numpy_time = timeit("matrix_numpy(numpy_matrix1, numpy_matrix2)", globals=globals(), number=100)

print(f"Pure Python addition : {python_time:.6f} seconds per loop")
print(f"Numpy addition : {numpy_time:.6f} seconds per loop")
```
