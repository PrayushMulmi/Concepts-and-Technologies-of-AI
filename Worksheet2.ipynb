Problem - 3:

The ’Embarked’ column in the Titanic dataset contains categorical data representing the ports of embarkation:

• ’C’ for Cherbourg
• ’Q’ for Queenstown
• ’S’ for Southampton

1. Use one-hot encoding to convert the ’Embarked’ column into separate binary columns (’Embarked C’, ’Embarked Q’, ’Embarked S’).

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Concepts and Technologies of AI/dataset/Titanic-Dataset.csv")
one_hot_encoding = pd.get_dummies(df['Embarked'], prefix='Embarked')
one_hot_encoding
```

2. Add these new columns to the original DataFrame.

```python
combined_columns = pd.concat([df, one_hot_encoding], axis=1)
print(combined_columns)
```

3. Drop the original ’Embarked’ column.

```python
df_without_embarked = df.drop(columns=['Embarked'])
print(df_without_embarked)
```

Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visualization to show how the survival distributions vary by gender.

```python
df_grouped_by_sex = df.groupby('Sex')
mean_survival_rate = df_grouped_by_sex['Survived'].mean()
print("Survival rate grouped by Sex : \n", mean_survival_rate)

mean_survival_rate.plot.bar()
```

Problem - 5:

Draw a visualization that breaks your visualization from Exercise 3 down by the port of embarkation (’Embarked’). In this instance, compare the ports ’C’ (Cherbourg), ’Q’ (Queenstown), and ’S’ (Southampton).

```python
import matplotlib.pyplot as plt
df.groupby(['Sex', 'Embarked'])['Survived'].mean().unstack().plot(kind='bar')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rate in accordance to Sex and Embarkation.')
plt.show()
```
