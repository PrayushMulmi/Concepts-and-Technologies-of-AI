```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Concepts and Technologies of AI/dataset/houseprice.csv")
df.head()
```

```python
house_age = df["HouseAge"].to_numpy()
house_floor = df["HouseFloor"].to_numpy()
house_area = df["HouseArea"].to_numpy()
house_price = df["HouseArea"].to_numpy()
```

```python
x0 = np.ones(len(house_age))
x2 = np.array([x0, house_age, house_floor, house_area]).T
W = np.array([0, 0, 0, 0])
y2 = np.array(house_area)
```

```python
def cost_function(X, Y, W):
  """
    Parameters:
    This function finds the Mean Square Error.
    Input Parameters:
    X : Feature Matrix
    Y : Target Matrix
    W : Weight Matrix
    Output Parameters:
    J : Mean Square Error
  """
  m = len(Y)
  J = 0
  h = np.dot(X, W)
  J = (1/2*m) * np.sum(np.square(h-Y))
  return J
```

```python
X_test = np.array([[1, 2], [3, 4], [5, 6]])
Y_test = np.array([1, 2, 3])
W_test = np.array([1, 2])

cost = cost_function(X_test, Y_test, W_test)

if cost == 0:
  print("Proceed further")
else:
  print("Something went wrong : Reimplement the cost function.")
  print("Cost function output : ", cost)
```

```python
initial_cost = cost_function(x2, y2, W)
print("Initial Cost : ", initial_cost)
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
  cost_history = [0] * iterations
  m = len(Y)
  for iteration in range(iterations):
    Y_pred = X.dot(W)
    loss = Y_pred - Y
    dW = X.T.dot(loss) / m
    W = W - alpha * dW
    cost = cost_function(X, Y, W)
    cost_history[iteration] = cost
  return W, cost_history
```

```python
alpha = 0.00000001
new_weight, cost_history = gradient_descent(x2, y2, W, alpha, 10000)
print("Final Weight : ", new_weight)
print("Final cost : ", cost_history[-1])
```

```python
def rmse(Y, Y_pred):
  """
    Parameters:
    This function finds the Root Mean Square Error.
    Input Parameters:
    Y : Target Matrix
    Y_pred : Predicted Matrix
    Output Parameters:
    rmse : Root Mean Square Error
  """
  rmse = np.sqrt(sum((Y - Y_pred) ** 2) / len(Y))
  return rmse
```

```python
def r2(Y, Y_pred):
  """
    Parameters:
    This function finds the R2 Score.
    Input Parameters:
    Y : Target Matrix
    Y_pred : Predicted Matrix
    Output Parameters:
    r2 : R2 Score
  """
  mean_y = np.mean(Y)
  ss_tot = sum((Y - mean_y) ** 2)
  ss_res = sum((Y - Y_pred) ** 2)
  r2 = 1 - (ss_res / ss_tot)
  return r2
```

```python
Y_pred = x2.dot(new_weight)
print("Root Mean Square Error : ", rmse(y2, Y_pred))
print("R^2 Value : ", r2(y2, Y_pred))
```
